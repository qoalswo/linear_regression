## 서울시 부동산 실거래가 정보

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import math
```

```python
df = pd.read_csv('./data.csv', encoding = 'cp949')
```

```python
df
```

```python
df = df[df['자치구명']=='강남구']
mydf = df[['접수연도', '물건금액(만원)', '건물면적(㎡)']]
mydf
```

```python
dl = []
for idx in mydf.index:
    dl.append(int(mydf.loc[idx, '물건금액(만원)'] / mydf.loc[idx, '건물면적(㎡)']))
mydf = mydf.copy()
mydf['면적당가격'] = dl
```

```python
mydf
```

```python
x_val = mydf['접수연도']
y_val = mydf['면적당가격']
#y_val = mydf['물건금액(만원)']

plt.scatter(x_val, y_val)
plt.show()
```

```python
line_filtter = LinearRegression()
line_filtter.fit(x_val.values.reshape(-1,1), y_val)

plt.scatter(x_val, y_val, c='green')
plt.plot(x_val,line_filtter.predict(x_val.values.reshape(-1,1)), c='red')
```

```python
predict_year = [2024, 2025, 2026, 2027, 2028]
for y in predict_year:
    print(f'강남구의 {y}년의 면적(㎡)당 가격 예측: {math.floor(line_filtter.predict([[y]])[0])}만 원')
```
